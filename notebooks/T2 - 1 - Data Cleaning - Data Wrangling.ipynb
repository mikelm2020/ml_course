# Data Wrangling - La cirugía de los datos

El data wrangling, a veces denominada data munging, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intención de hacerlo más apropiado y valioso para una variedad de propósitos posteriores, como el análisis. Un data wrangker es una persona que realiza estas operaciones de transformación.

Esto puede inlcluir munging, visualización de datos, agregacipon de datos, entrenamiento de un modelo estadístico, así como muchos otros usos potenciales.
La oscilación de datos como proceso generalmente sigue un conjunto de pasos generales que comienzan extrayendo los datos en forma cruda del origen de datos, dividiendo los datos en bruto usando algoritmos (por ejemplo clasificación) o analizando los datos en estrfucturas de datos predefinidas, y finalmente despositando el contenido rsultante en un sistema de almacenamiento (o sitio) para su uso futuro.

### Función que genera una lista de n números aleatorios enteros dentro del intervalo (a,b)

```python
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
        
```
